# Paginate bookmark page

```python
df = pd.read_csv('data/full_anime_info.csv')
df.head()
```

```python
def paginate(df, page, per_page=10):
# https://www.reddit.com/r/flask/comments/fxtfzm/is_there_a_way_to_add_pagination_to_my_webapp/
    count = len(df)
    # page = request.args.get('page', 1, type=int)
    per_page = 10 # define how many results you want per page
    pages = count // per_page # this is the number of pages
    offset = (page-1)*per_page # offset for SQL query
    return slice(offset, offset + per_page)
```

```python
slice_range = paginate(df, page=2)
```
